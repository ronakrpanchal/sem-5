```python
def calculator(a, b, c):
    if c == '+':
        return a + b
    elif c == '-':
        return a - b
    elif c == '*':
        return a * b
    elif c == '/':
        return a / b

calculator(1,2,'+')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path='/Users/ronakpanchal/Desktop/Finance Tracker/FT3/YourFinance.csv'

df=pd.read_csv(file_path)
df.dropna(inplace=True,axis=0)

expense=df[df['Amount']<0]
expense.iloc[:,df.columns.get_loc('Amount')]=abs(expense.iloc[:,df.columns.get_loc('Amount')])

grouped_expense=expense.groupby(['Category'])['Amount'].sum().reset_index()

plt.bar(x=grouped_expense['Category'],data=grouped_expense,height=grouped_expense['Amount'])
```

```python
plt.pie(x=grouped_expense['Amount'],labels=grouped_expense['Category'],autopct='%1.1f%%')
```
